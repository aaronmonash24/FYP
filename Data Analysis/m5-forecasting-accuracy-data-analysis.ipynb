```python
def categorize_item(item_id):
    if 'HOBBIES' in item_id:
        return 'HOBBIES'
    elif 'FOOD' in item_id:
        return 'FOODS'
    elif 'HOUSEHOLD' in item_id:
        return 'HOUSEHOLDS'
    else:
        return 'OTHER'
```
